# OVERVIEW

This is a toy data set derived from the General Social Survey that can be used for several purposes. First, the battery of abortion items can serve as toy data to illustrate mixed effects modeling as equivalent to a one-parameter (Rasch) model. Second, I include some covariates to also do some basic regressions. I think abortion opinions are useful learning tools for statistical inference for college students. Third, there's a time-series component as well for understanding how abortion attitudes have changed over time.

**This dataset consits of total 18 column and 64,814 row entries. Columns are as follow:**

1) **id**: a unique respondent identifier

2) **year**: the survey year

2) **age**: the respondent's age in years

4) **race**: the respondent's race, as character variable

5) **sex**: the respondent's gender, as character variable

6) **hispaniccat**: the respondent's Hispanic ethnicity, as character variable

7) **educ**: how many years the respondent spent in school

8) **partyid**: the respondent's party identification, as character variable

9) **relactiv**: the self-reported religious activity of the respondent on a 1:11 scale

10) **abany**: a binary variable that equals 1 if the respondent thinks abortion should be legal for any reason. 0 indicates no                support for abortion for any reason.

11) **abdefect**: a numeric vector that equals 1 if the respondent thinks abortion should be legal if there is a serious defect                   in the fetus. 0 indicates no support for abortion in this circumstance.

12) **abnomore**: a numeric vector that equals 1 if the respondent thinks abortion should be legal if a woman is pregnant but                     wants no more children. 0 indicates no support for abortion in this circumstance.

13) **abhlth**: a numeric vector that equals 1 if the respondent thinks abortion should be legal if a pregnant woman's health is                 in danger. 0 indicates no support for abortion in this circumstance.

14) **abpoor**: a numeric vector that equals 1 if the respondent thinks abortion should be legal if a pregnant woman is poor and                 cannot afford more children. 0 indicates no support for abortion in this circumstance.

15) **abrape**: a numeric vector that equals 1 if the respondent thinks abortion should be legal if the woman is a rape victim.                 0 indicates no support for abortion in this circumstance.

16) **absingle**: a numeric vector that equals 1 if the respondent thinks abortion should be legal if a pregnant woman is single                   and does not want to marry the man who impregnated her. 0 indicates no support for abortion in this                             circumstance.

17) **pid**: partyid recoded so that 7 = NA

18) **hispanic**: a dummy variable that equals 1 if the respondent is any way Hispanic

```python
# 2.Create a line plot to visualize the trend in abortion opinions over the years.

avg_opinion_by_year = df.groupby('year')['abany'].mean()
plt.plot(avg_opinion_by_year.index, avg_opinion_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Opinion on Abortion')
plt.title('Trend in Abortion Opinions over the Years')
plt.show()
```

#Conclusion

From the line plot visualizing the trend in abortion opinions over the years, we can draw the following conclusions:

Changing attitudes: The plot provides insights into the evolving opinions on abortion over time. By examining the trend line, we can observe whether support for abortion has increased, decreased, or remained relatively stable over the years.

Impact of time: The plot helps us understand the influence of time on abortion opinions. It allows us to identify any significant shifts or patterns in support for abortion across different years.

```python
# 3.What is the distribution of years of education ('educ') among the respondents?

education_distribution = df['educ'].value_counts().sort_index()
print("Distribution of Years of Education:")
print(education_distribution)
```

```python
# 4.Create a stacked area chart to visualize the changing proportions of abortion opinions over the years.

yearly_proportions = df.groupby('year')[['abany', 'abdefect', 'abnomore', 'abhlth', 'abpoor', 'abrape', 'absingle']].mean()
plt.stackplot(yearly_proportions.index, yearly_proportions.values.T, labels=yearly_proportions.columns)
plt.xlabel('Year')
plt.ylabel('Proportion')
plt.title('Changing Proportions of Abortion Opinions over the Years')
plt.legend(loc='upper left')
plt.show()
```

```python
# 6.Generate a pie chart representing the percentage of respondents supporting, opposing, and unsure about abortion.

df['abany'].fillna('unsure', inplace=True)

support_count = df[df['abany'] == 1].shape[0]
oppose_count = df[df['abany'] == 0].shape[0]
unsure_count = df[df['abany'] == 'unsure'].shape[0]

total_count = support_count + oppose_count + unsure_count

support_percentage = (support_count / total_count) * 100
oppose_percentage = (oppose_count / total_count) * 100
unsure_percentage = (unsure_count / total_count) * 100

labels = ['Support', 'Oppose', 'Unsure']
sizes = [support_percentage, oppose_percentage, unsure_percentage]
colors = ['green', 'red', 'yellow']
explode = (0.1, 0, 0)  # Explode the first slice (support)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

plt.title('Opinions on Abortion')

plt.axis('equal')

plt.show()
```

```python
# 15.Generate a grouped bar chart to compare the support for abortion for any reason among different Hispanic ethnicities.

support_abortion = df[df['abany'] == 1]

ethnicity_counts = support_abortion.groupby('hispaniccat').size()

ethnicities = ethnicity_counts.index.tolist()
count_values = ethnicity_counts.tolist()

x = range(len(ethnicities))

bar_width = 0.35

plt.bar(x, count_values, width=bar_width, label='Support for Abortion')

plt.xticks(x, ethnicities)

plt.xlabel('Hispanic Ethnicity')
plt.ylabel('Number of Respondents')
plt.title('Support for Abortion for Any Reason by Hispanic Ethnicity')
plt.legend()

plt.show()
```

```python
# 17.Create a stacked bar chart comparing the opinions on abortion between different education levels.

support_abortion = df[df['abany'] == 1]

education_counts = support_abortion.groupby('educ').size()

education_opinion_counts = support_abortion.groupby('educ')[['abdefect', 'abnomore', 'abhlth', 'abpoor', 'abrape', 'absingle']].sum()

abortion_opinions = ['abdefect', 'abnomore', 'abhlth', 'abpoor', 'abrape', 'absingle']

x = range(len(education_counts))

bar_width = 0.8

bottom = [0] * len(education_counts)
for opinion in abortion_opinions:
    plt.bar(x, education_opinion_counts[opinion], width=bar_width, bottom=bottom, label=opinion)
    bottom += education_opinion_counts[opinion]

plt.xticks(x, education_counts.index)

plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.title('Opinions on Abortion by Education Level')

plt.legend()

plt.show()
```
